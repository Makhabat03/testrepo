```python
import numpy as np
```

```python
x = np.array([i for i in range(1000)]) # We have one feature x and we have 1000 samples, m = 1000 and n = 1
y =  x # y = 1*x + 0  is the true value
```

```python
# check
print (len(x), len(y))
```

```python
# Assume that your hypotheis is theta_1 = 0.99 and theta_0 = 0

y_hat = h_theta = 0.99 * x + 0

# y_hat contains predictions for all 1000 samples
```

```python
#check
len(y_hat)
```

```python
# Breakdown of the parts of loss calculation
print (np.square(y_hat-y).shape)
print ((y_hat-y).shape)
print (np.sum(np.square(y_hat-y)))
```

```python
# calculate cost (also called loss)

#j_theta = cost = loss = (1/2) * (1/len(x)) * np.sum(np.square(y_hat-y))  # This calculation of j_theta include the 1/2 term. YOu will get same results either way

j_theta = cost = loss =  (1/len(x)) * np.sum(np.square(y_hat-y))
```

```python
print (j_theta)
```

In the cells above we have 1 hypothesis: theta_1 = 0.99 and theta_0 = 0

What if we generated multiple hypothesis by varying theta_1?

Lets generate multiple hypothesis and calculate J_theta for every hypothesis

```python
from random import uniform

all_theta_1 = [] # list to store theta_1 for every new hypothesis
all_j_theta = [] # list to store cost function for each hypothesis
for i in range(1000):
  theta_1 = uniform(-5,5)  # varying theta_1 from -5 to 5
  all_theta_1.append(theta_1)
  y_hat = theta_1 * x  + 0 # Remember we are setting theta_0 = 0
  #j_theta = (1/2) * (1/len(x)) * np.sum(np.square(y_hat-y))
  j_theta = (1/len(x)) * np.sum(np.square(y_hat-y))  # not including 1/2 does not make a difference. Like i said 1/2 is for mathematical convenience
  all_j_theta.append(j_theta)

print (len(all_theta_1))
print (len(all_j_theta))
```

```python
import matplotlib.pyplot as plt
# graph to plot x vs y (true value)
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Plot J_theta vs. theta_1
plt.plot(all_theta_1,all_j_theta,'bo')
plt.xlabel('theta_1')
plt.ylabel('Cost')
plt.show()
```

As seen above J_Theta or cost is lowest when theta_1 is approximately = 1

Using the mean squared error as the cost function allows us to pick the hypotheis for which the cost (or J_Theta) is the lowest.

```python
import random
```

```python
x = np.array([random.randint(1,10000) for i in range(10000)])

y = 3 * x

print (len(x), len(y))
```

```python
def mean_square_error(y,y_hat):
  return ((1/2) * (1/len(y)) * np.sum(np.square(y_hat-y)))
```

```python
from random import uniform

def hypothesis_generator(start_theta_1, end_theta_1, x):
  all_theta_1 = []
  all_j_theta = []
  for i in range(len(x)):
    theta_1 = uniform(start_theta_1,end_theta_1)
    all_theta_1.append(theta_1)
    y_hat = theta_1 * x    # calculate y_hat
    #j_theta = (1/2) * (1/len(x)) * np.sum(np.square(y_hat-y))
    j_theta = mean_square_error(y,y_hat) # calculate j_theta
    all_j_theta.append(j_theta)

  return all_theta_1, all_j_theta
```

```python
all_theta_1, all_j_theta = hypothesis_generator(0,6,x)
```

```python
plt.plot(x,y,'r1')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(all_theta_1,all_j_theta,'b1')
plt.xlabel('theta_1')
plt.ylabel('Cost')
plt.show()
```
